# Black-Scholes Option Greeks
Call price and various greeks over time and with respect to changes in volatility

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

N=norm.cdf
exp = np.exp
sqrt=np.sqrt
log=np.log
pi=np.pi
```

### Sigma = volatility, r = risk-free rate, K = strike price, S = current stock price, t = time to expiration

```python
sigma=0.3 
r=0.01
K=10
S=20
t=1
d1=1/(sigma*sqrt(t))*(log(S/K)+(r+sigma**2/2)*t)
d2=d1-sigma*sqrt(t)
```

### Black-Scholes call option formula

```python
call = N(d1)*S-N(d2)*K*exp(-r*t)
```

```python
print('Delta:',N(d1))
print('Call price:', call)
```

## Call price over time
Price decreases with time and goes to zero at expiration

```python
sigma=0.3
r=0.01
K=10
S=10
t=np.linspace(0.01, 1, 100)
d1=1/(sigma*sqrt(t))*(log(S/K)+(r+sigma**2/2)*t)
d2=d1-sigma*sqrt(t)
call = N(d1)*S-N(d2)*K*exp(-r*t)
plt.scatter(1-t,call)
plt.show()
```

## Call price change as interest rate changes

```python
sigma=0.3
r=np.linspace(0, 0.2, 100)
K=10
S=10
t=1
d1=1/(sigma*sqrt(t))*(log(S/K)+(r+sigma**2/2)*t)
d2=d1-sigma*sqrt(t)
call = N(d1)*S-N(d2)*K*exp(-r*t)
plt.scatter(r,call)
plt.show()
```

## Gamma (ATM) over time

```python
sigma=0.3
r=0.01
K=10
S=10
t=np.linspace(0.01, 1, 100)
d1=1/(sigma*sqrt(t))*(log(S/K)+(r+sigma**2/2)*t)
gamma=exp(-0.5*d1**2)/(sqrt(2*pi)*S*sigma*sqrt(t))
plt.scatter(1-t,gamma)
plt.show()
```

# Vega over time

```python
sigma=0.3
r=0.01
K=10
S=10
t=np.linspace(0.01, 1, 100)
vega=exp(-(d1**2/2))/(2*pi)**0.5*S*sqrt(t)/100
plt.scatter(t,vega)
plt.show()
```

# Rho over time

```python
sigma=0.3
r=0.01
K=10
S=10
t=np.linspace(0.01, 1, 100)
d1=1/(sigma*sqrt(t))*(log(S/K)+(r+sigma**2/2)*t)
d2=d1-sigma*sqrt(t)
rho=S*t*exp(-r*t)*N(d2)/100
plt.scatter(t,rho)
plt.show()
```

# Theta changes as volatility changes

```python
sigma=np.linspace(0.01, 1, 100)
r=0.01
K=10
S=10
t=1
d1=1/(sigma*sqrt(t))*(log(S/K)+(r+sigma**2/2)*t)
d2=d1-sigma*sqrt(t)
theta=((-exp(-0.5*d1**2))/sqrt(2*pi)*(S*sigma/(2*sqrt(t)))-(r*K*exp(-r*t)*N(d2)))/365
plt.scatter(sigma,theta)
plt.show()
```
